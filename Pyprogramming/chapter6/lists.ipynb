```python
# Empty list instantiation
colors = []
# Get user's input
color = input("Enter you favouring colors (3): ")

# Appending user's input to the created empty list
colors.append(color)

# Print the results
print(f"Your favourite colors are: {colors}")
```

```python
# collections
colors = []

for i in range(3):
    color = input(f"Enter your favourite color #{1 + i}: ")
    colors.append(color)

# Priting the results
print(f"Your favourite colors are: {colors}")
```

Write a program that asks the user for 5 numbers and print only the even numbers

```python
numbers = []
for i in range(5):
    number = input(f"Enter number #{i + 1}: ")

    # Check if the input number is an even number
    try:
        if not (int(number) % 2):
            numbers.append(number)
    except Exception as err:
        print(f"Invalid. This error has occured: {err}")

# Printing the results (even numbers only)
print(f"Even numbers are: {numbers}")
```

```python
numbers = []
for i in range(5):
    number = input(f"Enter number #: {i + 1}: ")

    # Validate the input number
    while True:
        try:
            number = int(number)
            
            # Appending the number to the list
            numbers.append(number)

            # Exciting loop if valid
            break
        except Exception as err:
            print("Invalid number. Try again!")

print(f"Your numbers are: {numbers}")

# for i in numbers[:]:
#     if i % 2:
#         numbers.remove(i)

# We can as well use a list comprehension
numbers = [number for number in numbers[:] if number % 2 == 0]

# Printing the new list
print(f"Even number list: {numbers}")

```

```python
# Inserting odd numbers in from of a list
numbers = []

for i in range(5):
    while True:
        number = input(f"Enter number #{i + 1}: ")
        try:
            number = int(number)
            numbers.append(number)

            break
        except Exception as err:
            print(err)
for i in numbers[:]:
    if i % 2:
        # Removing and inserting the odd number
        numbers.remove(i)
        numbers.insert(0, i)

print(f"New list: {numbers}")
```

```python
# Creating a list of 6 numbers from input
numbers = [int(input(f"Enter number #{i + 1}: ")) for i in range(6)]

# printing the first and last element
print(f"The first and last : {numbers[0], numbers[-1]}")

# The middle four elements
totalNumbers = len(numbers)
if totalNumbers < 4:
    print("Fewer than four elements")

elif totalNumbers == 4:
    print(f"The middle 4 elements: {numbers}")
else:
    if totalNumbers % 2 == 0:
        midNumber = totalNumbers // 2
        start = midNumber - 2
        end = midNumber + 2
        print(f"The middle 4 elements: {numbers[start: end]}")
    else:
        print("You can not get the middle four elements!")
# Reversed list
print(f"Reversed list: {numbers[::-1]}")

# Getting every second number
print(f"Every second number: {numbers[::2]}")
```
