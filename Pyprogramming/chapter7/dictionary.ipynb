# Dictionaries

Accessing car information safely

```python
# Car dictionary
car = {
    "brand": "Toyota",
    "year": 2020
}
```

Use `get()` to safely print the brand of the car

```python
car_brand = car.get("brand")
print(f"Brand: {car_brand}")
```

Use `get()` to safely print the color of the car. Color does not exist yet, so it should not crush

```python
car_color = car.get("color")
print(f"Color: {car_color}")
```

Update `get()` to make sure it prints `unknown` if the color does not exists

```python
car_updated_color = car.get("color", "unknown")
print(f"Color: {car_updated_color}")
```

## The `setdefault()`

It ensures the key exists and can set a default value if the key is missing.

```python
car.setdefault("color", "black")
print(f"car: {car.keys()}")
print(f"Color: {car["color"]}")
print(car)
```

## Compute frequency using `setdefault()`

```python
# Computing frequency
message = "stockout"

# Counter dictionary
count = {}

# Loop 
for character in message:
    
    # Default set (C_x)
    c_x = count.setdefault(character, 0)
    
    # Compute frequency
    count[character] += 1

# Print results
print(count)
```

```python
# Dictionary items
for key, value in count.items():
    print(key + ":", value)
```

## Nested dictionaries

```python
cars = {
    "Toyota": {"year": 2020, "color": "red"},
    "Honda": {"year": 2019, "color": "blue"},
    "Nissan" : {"year": 2021, "color": "black"}
}

# Print cars
print(cars)
```

```python
# Q1: Print all car names in the dictionary
car_names = cars.keys()

print(list(car_names))
```

```python
# Q2: Print the year of the toyota
toyota_year = cars["Toyota"]["year"]

print(toyota_year)
```

```python
# Add a new car: "Ford" with year: 2018 and color: green

# Step 1. Set the key
cars.setdefault("Ford", {"year": 2018, "color": "green"})

# Print
print(cars)
```
